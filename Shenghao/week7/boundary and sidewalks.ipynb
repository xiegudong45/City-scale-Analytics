```python
import geopandas as gpd
import numpy as np
import os.path
```

```python
boundary_dir = '../data/Park Boundary.geojson'
```

```python
boundary_df = gpd.read_file(boundary_dir)
boundary_df.head()
```

```python
boundary_df = boundary_df.replace(to_replace='None', value=np.nan).dropna()
boundary_df.head()
```

```python
boundary_df['name'].apply(lambda x: x.upper())
boundary_df.head()
```

```python
boundary_df = boundary_df[boundary_df['name'].str.contains('PARK')]
boundary_df.head()
```
